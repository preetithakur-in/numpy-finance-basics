```python
'''Q4 Given this array of daily closing prices:
prices = np.array([100, 105, 103, 110, 108, 115])
Calculate the daily return for each day using: '''

prices = np.array([100, 105, 103, 110, 108, 115])
daily_returns = (prices[1:] - prices[:-1])/ prices[:-1]
print(np.round(daily_returns, 3))
```

```python
'''Q5:- From the same prices array, calculate the standard deviation of daily returns.'''
prices = np.array([100, 105, 103, 110, 108, 115])
returns = (prices[1:] - prices[:-1])/ prices[:-1]
std_dev = np.std(returns)
print("Daily Returns:", np.round(returns, 3))
print("Standard Deviation of Returns:", std_dev)
```

```python
'''Q7. Given daily returns:
returns = np.array([0.01, -0.02, 0.015, 0.03, -0.01])
Calculate the cumulative return using np.cumprod().'''
prices = np.array([0.01, -0.02, 0.015, 0.03, -0.01])
returns = (prices[1:] - prices[:-1])/ prices[:-1]
cumulative_returns = np.cumprod(1 + returns)
print("Daily_Returns:", np.round(returns, 3))
print("cumulative_returns", cumulative_returns)
```

```python
'''Q8:- Generate 252 random daily returns with mean 0.001 and std 0.02, the annualized volatility.
annualized vol = daily std * v252'''
mean = 0.001
std = 0.02
num_days = 252
np.random.seed(42)
daily_returns = np.random.normal(loc=mean, scale=std, size=num_days)
daily_std = np.std(daily_returns)
annualized_vol = daily_std * np.sqrt(252)

print(f"Daily std (volatility): {daily_std:.3f}")
print(f"Annualized Volatility: {annualized_vol:.4f}")
```

```python
'''Q9:- Given:
prices = np.array([100, 105, 103, 110, 108, 115, 120, 118, 125, 130])
Calculate the 5-day moving average using a loop and Numpy slicing.'''
prices = np.array([100, 105, 103, 110, 108, 115, 120, 118, 125, 130])
window_size = 5
moving_averages = []

for i in range(len(prices)- window_size + 1):
    window = prices[i : i + window_size]
    window_average = np.sum(window)/window_size
    moving_averages.append(window_average)

print(np.array(moving_averages))
    
```

```python
'''Q10:- Given:
stock_a = np.array([0.01, 0.02, -0.01, 0.03])
stock_b = np.array([0.02, 0.01, 0.015, -0.01])
stock_c = np.array([-0.01, 0.03, 0.02, 0.01]) 
Stack them into a 2D array and calculate the average return per stock in one line.'''
stock_a = np.array([0.01, 0.02, -0.01, 0.03])
stock_b = np.array([0.02, 0.01, 0.015, -0.01])
stock_c = np.array([-0.01, 0.03, 0.02, 0.01]) 

average_returns = np.vstack([stock_a, stock_b, stock_c]).mean(axis=1)

print(average_returns)
```
